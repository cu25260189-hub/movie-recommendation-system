```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Creating a simple movie dataset
movies = {
    "title": [
        "Inception",
        "Interstellar",
        "The Dark Knight",
        "Avengers",
        "Titanic",
        "Joker",
        "Iron Man",
        "The Matrix"
    ],
    "details": [
        "dream mind heist thriller",
        "space time science fiction",
        "batman crime action thriller",
        "superheroes team action sci fi",
        "love romance ship tragedy",
        "psychological crime drama",
        "technology superhero marvel",
        "virtual reality artificial intelligence"
    ]
}

movie_df = pd.DataFrame(movies)

# Converting movie descriptions into numbers
vectorizer = TfidfVectorizer()
movie_vectors = vectorizer.fit_transform(movie_df["details"])

# Finding similarity between movies
similarity = cosine_similarity(movie_vectors)

# Function to recommend movies
def recommend(movie_name):
    if movie_name not in movie_df["title"].values:
        print("Sorry, movie not found.")
        return

    movie_index = movie_df[movie_df["title"] == movie_name].index[0]
    scores = list(enumerate(similarity[movie_index]))

    # Sorting movies based on similarity
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nMovies similar to '{movie_name}':\n")

    count = 0
    for i in scores:
        if count == 3:
            break
        if i[0] != movie_index:
            print("- " + movie_df.iloc[i[0]]["title"])
            count += 1


# Running the program
user_movie = input("Enter a movie name: ")
recommend(user_movie)
```
